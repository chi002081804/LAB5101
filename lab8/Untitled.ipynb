```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('./diabetes.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data.describe()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8,6))
sns.histplot(data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()  # Display the plot
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='BMI', y='Glucose', hue='Outcome', data=data, palette='viridis')
plt.title('BMI vs. Glucose Scatter Plot')
plt.xlabel('BMI')
plt.ylabel('Glucose Level')
plt.legend(title='Outcome', labels=['No Diabetes', 'Diabetes'])
plt.show()  
```

```python
corr_matrix = data.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()  # Display the plot
```

```python
plt.figure(figsize=(8,6))
ax = sns.boxplot(x='Outcome', y='BloodPressure', data=data, hue='Outcome', palette='Set2')
plt.title('Blood Pressure by Diabetes Outcome')
plt.xlabel('Outcome (0 = No Diabetes, 1 = Diabetes)')
plt.ylabel('Blood Pressure')
ax.legend_.remove()
plt.show()
```
